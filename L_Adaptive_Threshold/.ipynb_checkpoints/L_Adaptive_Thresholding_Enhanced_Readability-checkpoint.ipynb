- An improved image thresholding technique where the threshold value is varied at each pixel. A common technique is to use the average intensity in a neighbourhood to set the threshold.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def show(img_BGR,title='RGB_img'):
    img_RGB = cv2.cvtColor(img_BGR,cv2.COLOR_BGR2RGB)
    plt.imshow(img_RGB)
    plt.title(title)
    plt.show()
```

```python
pp = cv2.imread('bookpage.jpg')
show(pp)
```

### static threshold

```python
pp_gray  = cv2.cvtColor(pp,cv2.COLOR_BGR2GRAY)
_,pp_static = cv2.threshold(pp_gray,15,255,cv2.THRESH_BINARY)
plt.imshow(pp_static,cmap='gray')
plt.show()
```

### Adaptive Threshold

```python
pp_adaptive = cv2.adaptiveThreshold(pp_gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,71,1)
plt.imshow(pp_adaptive,cmap='gray')
plt.show()
```
