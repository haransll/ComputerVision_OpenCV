### matplotlib: why--> to produce inline plots

```python
#note: matplotlib stores RGB whereas cv2 stores images as BGR (color correction)

imgRGB = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(imgRGB, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
# plt.plot([200,300,400],[100,200,300],'c', linewidth=5)
plt.show()
```

## loading and saving video feed

- view from video file or camera in RT
- play
- save as a list of frames or mp4

```python
fname='/Users/HARAN/Library/Mobile Documents/com~apple~CloudDocs/CV/project/p1_train_spotting/help_trainspotting'
videofile = cv2.VideoCapture(fname+'/video/orig.mp4')
```

```python
ret,orig_frame = videofile.read()

gray_frames = []
while ret:
    
    
    
    gray_frame = cv2.cvtColor(orig_frame, cv2.COLOR_BGR2GRAY)
    cv2.imshow('gray_frame',gray_frame)
    
    
    
    #save
    #gray_frames.append(gray_frame)

    cv2.imshow('orig_frame',orig_frame)
    ret,orig_frame = videofile.read()
    
    cv2.waitKey(1000)
    
    if cv2.waitKey(1) & 0xFF ==ord('q'):
        break
        
cv2.destroyAllWindows()
cv2.waitKey(1)        
```
